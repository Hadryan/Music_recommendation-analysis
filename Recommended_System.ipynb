##  Song Recommender System

#### Import packages and get csv file

```python
# Get Csv File
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
import time
from sklearn.externals import joblib
song_df = pd.read_csv(r'C:\Users\Nathan\Documents\music recommendation system\song_df.csv')
import recomendar as recomendar
```

```python
# Show data
song_df.head(10)
```

```python
song_df = song_df.head(100000)
```

```python
song_df['song'] = song_df['title'].map(str) + " - " + song_df['artist_name']
```

```python
users = song_df['user_id'].unique()
```

### Creating song recommender by user, by item similarity based collaborative filtering

```python
train_data, test_data = train_test_split(song_df, test_size = 0.20, random_state=0)
```

```python
#Create an instance of item similarity based recommender class
is_model = recomendar.item_similarity_recommender_py()
is_model.create(train_data, 'user_id', 'song')
```

```python
#Use the personalized model to make some song recommendations
user_id = users[5]
user_items = is_model.get_user_items(user_id)
#
print("------------------------------------------------------------------------------------")
print("Training data songs for the user userid: %s:" % user_id)
print("------------------------------------------------------------------------------------")

for user_item in user_items:
    print(user_item)

print("----------------------------------------------------------------------")
print("Recommendation process going on:")
print("----------------------------------------------------------------------")

#Recommend songs for the user using personalized model
is_model.recommend(user_id)
```
